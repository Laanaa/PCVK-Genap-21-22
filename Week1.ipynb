```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
urls = ["https://cdn1-production-images-kly.akamaized.net/ahxsQdSm2Pmkdk5KxuUuq4yQEsg=/1280x720/smart/filters:quality(75):strip_icc():format(webp)/kly-media-production/medias/3341793/original/004647700_1609922345-red-fox-5456627_1920.jpg",
        "https://i0.wp.com/rimbakita.com/wp-content/uploads/2020/10/gambar-kartun-doraemon-768x432.jpg",
        "https://ae01.alicdn.com/kf/HTB1JZ9SLXXXXXcEXVXXq6xXFXXXH/Simulasi-Rusa-Polyethylene-Bulu-Rusa-Model-Hadiah-Lucu-Sekitar-11-Cm-X-4-Cm-X-14.jpg"]

for url in urls :
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat ((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
kedalaman = image_2.shape[2]
tinggi2,lebar2,kedalaman2 = np.shape(image_2)
print("resolusi image : tinggi x lebar x kedalaman = ", tinggi," x ",lebar , "x" ,kedalaman)
print("resolusi image 2 : tinggi x lebar x kedalaman = ", tinggi2," x ",lebar2 , "x" ,kedalaman2)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#horizontal 
for y in range (lebar) :
  #vargambar[koordinat x, koordinat y]
  image_3[int((tinggi)/2),y] = [123,252,253]

#vertikal
for y in range (tinggi) :
  image_3[y, int((lebar)/2)] = [123,252,253]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
